-> i want to calculate the integral of a random function between 0 1 

```python
def f(x): 
    return 2*x
```

```python
a = 0
b = 1

n_samples = 10000
random_points = np.random.uniform(a, b, n_samples)
print(random_points)
function_values = f(random_points)

integral_estimate = (b-a)* np.mean(function_values)

print("monte carlo estimate: "+ str(integral_estimate))
```

```python
def f3(x):
    # Linearly in the interval [0, 1]: f(x) = (x + 1)/2 and zero elsewhere. x is a random number between 0 and 1
    condition = (x >= -1) & (x <= 1)
    return np.where(condition,  2 * (np.sqrt(x))-1, 0) #the inverse of the cdf of the function asked

hist(f3(random_number), bins=50)
```

hit and miss method: 

```python
def f5(x): 
    return (x+1)/2

# Step 2: Define the domain and range
a, b = -1, 1  # Domain [0, pi]
f_min, f_max = f5(a), f5(b)  

# Step 3: Generate random points
n_samples = 100000
t_random = np.random.uniform(a, b, n_samples)
s_random = np.random.uniform(f_min, f_max, n_samples)

hits = []

# Step 4: Count the hits
for t, s in zip(t_random, s_random):
    if s<=f5(t):
        hits.append(t)
    
    

# Optional: Plot the function and the random points
hist(hits, bins=50)
```

```python
def f6(x): 
    return -x*(1-x)*np.exp(-x**2)*np.log(1-x)

# Step 2: Define the domain and range
a, b = 0, 1  
f_min, f_max = 0, 0.3  

# Step 3: Generate random points
n_samples = 100000
t_random = np.random.uniform(a, b, n_samples)
s_random = np.random.uniform(f_min, f_max, n_samples)

hits = []

# Step 4: Count the hits
for t, s in zip(t_random, s_random):
    if s<=f6(t):
        hits.append(t)
    
# Optional: Plot the function and the random points
hist(hits, bins=50)
```

Exercise 2: page 30 -> gaussian RNG

```python
import numpy as np

def polar_to_cartesian(r, theta):
    x = r * np.cos(theta)
    y = r * np.sin(theta)
    return x, y

# Generate random numbers from an exponential distribution
n_samples = 100000
random_numbers = np.random.exponential(scale=1, size=n_samples)

# Transform the random numbers to match the desired distribution
r = np.sqrt(2 * random_numbers)
theta = np.random.uniform(0, np.pi*2, n_samples)

x, y = polar_to_cartesian(r, theta)
# Plot the histogram of the generated random numbers
plt.hist(x, bins=50, density=True, alpha=0.5, color='g', label='Generated Distribution')
plt.xlabel('x')
plt.ylabel('Density')
plt.legend()
plt.show() 
```

If we want to have control on the mean and variance? 

```python
mean = 5
variance = 9
import numpy as np

def polar_to_cartesian(r, theta):
    x = r * np.cos(theta)
    y = r * np.sin(theta)
    return x, y

# Generate random numbers from an exponential distribution
n_samples = 100000
random_numbers = np.random.exponential(scale=1, size=n_samples) 

# Transform the random numbers to match the desired distribution
r = np.sqrt(2 * random_numbers)
theta = np.linspace(0, np.pi*2, n_samples)

x, y = polar_to_cartesian(r, theta)

x = mean + np.sqrt(variance) * x

# Plot the histogram of the generated random numbers
plt.hist(x, bins=50, density=True, alpha=0.5, color='g', label='Generated Distribution')
plt.xlabel('x')
plt.ylabel('Density')
plt.legend()
plt.show()
```
