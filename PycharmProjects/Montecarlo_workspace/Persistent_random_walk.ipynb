```python
import numpy as np
import matplotlib.pyplot as plt
import random
import math
```

```python
position  = 0 # initial position

alpha = 0.01 # probability

def random_step(position, alpha):
    if random.uniform(0, 1) > 1-2*alpha:
        if random.uniform(0, 1)>0.5:
            position += 1
        else:
            position -= 1
    
    return position

    
```

```python
def calculate_time_arriving_position(arriving_position, alpha, max_steps):
    position = 0
    time = 0
    while position != arriving_position and time < max_steps:
        position = random_step(position, alpha)
        time += 1
    return time

```

```python
number_of_simulations = 500
average_time = 0
max_steps = 1000000
times = []

for i in range(number_of_simulations):
    print(i/number_of_simulations)
    time = calculate_time_arriving_position(20, alpha,max_steps )
    times.append(time)
    average_time+=time

print(average_time/number_of_simulations)
plt.plot(times)

```

```python
#print the distribution of the time
plt.hist(times, bins = 100)
```

```python
def continuous_time_random_walk(arriving_position, alpha, max_steps):
    position = 0
    total_time = 0
    time_spent = []

    for _ in range(max_steps):
        # Time spent at the current position follows an exponential distribution
        time_at_position = np.random.exponential(1 / alpha)
        total_time += time_at_position
        time_spent.append(time_at_position)

        # Choose the direction of the jump with equal probability
        if random.uniform(0, 1) > 0.5:
            position += 1
        else:
            position -= 1

        # Check if the walker has reached the arriving position
        if position == arriving_position:
            break

    return total_time, time_spent
```

```python
# Parameters
arriving_position = 1000000
alpha = 0.01
max_steps = 1000000

# Run the simulation
total_time, time_spent = continuous_time_random_walk(arriving_position, alpha, max_steps)

# Calculate the average time spent at each lattice point
average_time_spent = np.mean(time_spent)
print(f"Total time: {total_time}")
print(f"Average time spent at each lattice point: {average_time_spent}")
```
